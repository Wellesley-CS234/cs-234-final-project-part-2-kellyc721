```python
import pandas as pd
```

```python
df = pd.read_csv("data/predicted_categories.csv")
df.head()
```

```python
df = df.drop(columns=['predicted_label', 'score'])
```

```python
df_views = pd.read_csv("data/covid_articles_matched_qids.csv")
```

```python
df_views.drop(columns=['country_code', 'project'])
```

```python
merged_df = df.merge(
    df_views[['date', 'article', 'pageviews']],
    on='article',
    how='left'
)
merged_df.dropna()
```

```python
merged_df.to_csv("data/categories_pageviews.csv")
```
