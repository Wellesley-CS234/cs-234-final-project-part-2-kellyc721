```python
import pandas as pd
from scipy.signal import find_peaks
import numpy as np
```

```python
df = pd.read_csv("data/covid_articles_matched_qids.csv")
```

```python
daily = df.groupby("date")["pageviews"].sum().reset_index()
```

```python
daily.head()
```

```python
max_views = daily['pageviews'].max()
print(max_views)
peak_indices, _ = find_peaks(
    daily['pageviews'], 
    prominence=0.01 * max_views # Threshold to find only major events
)
```

```python
peak_dates = daily.iloc[peak_indices][['date', 'pageviews']]
peak_dates.sort_values('date', ascending=True)
```

```python
peak_dates.to_csv("data/known_peaks.csv")
```
